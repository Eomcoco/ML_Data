# 2.인구 구조 시각화

### 2.1. 지역별 총 인구수 

```python
import pandas as pd
import matplotlib.pyplot as plt

# 모든 열을 문자열로 변환
data = data.astype(str)

# 행정구역 이름에서 괄호 및 숫자 제거
data['행정구역'] = data['행정구역'].str.replace(r'\s*\(.*?\)', '', regex=True)

# 인구 수 및 세대 수 전처리
data['2024년09월_총인구수'] = data['2024년09월_총인구수'].replace({'\\D': ''}, regex=True).str.replace(',', '').astype(int)
data['2024년09월_세대수'] = data['2024년09월_세대수'].replace({'\\D': ''}, regex=True).str.replace(',', '').astype(int)

# "전국" 행 제거
data = data[data['행정구역'] != '전국']

# 데이터 정렬
data_sorted = data.sort_values(by='2024년09월_총인구수', ascending=False)

# 지역별 총인구수 시각화
plt.figure(figsize=(12, 6))
plt.bar(data_sorted['행정구역'], data_sorted['2024년09월_총인구수'], color='skyblue')
plt.xlabel('행정구역')
plt.ylabel('총 인구수')
plt.title('지역별 총 인구수 (내림차순)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## 2.1. 성별 인구 수 시각화

```python
# 성별 인구 수 시각화
data['2024년09월_남자 인구수'] = data['2024년09월_남자 인구수'].str.replace(',', '').astype(int)
data['2024년09월_여자 인구수'] = data['2024년09월_여자 인구수'].str.replace(',', '').astype(int)

male_population = data['2024년09월_남자 인구수'].sum()
female_population = data['2024년09월_여자 인구수'].sum()

# 데이터 준비
labels = ['남자', '여자']
sizes = [male_population, female_population]
colors = ['lightblue', 'lightpink']

# 파이 차트 생성
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # 원을 동그랗게 유지
plt.title('성별 인구 비율')
plt.show()
```

```python
# 지역별 세대당 인구 수 시각화
data['세대당 인구 수'] = data['2024년09월_총인구수'] / data['2024년09월_세대수']

# 데이터 정렬
data_sorted = data.sort_values(by='세대당 인구 수', ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(data_sorted['행정구역'], data_sorted['세대당 인구 수'], color='lightgreen')
plt.xlabel('행정구역')
plt.ylabel('세대당 인구 수')
plt.title('지역별 세대당 인구 수 (내림차순)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 파이 차트 생성
plt.figure(figsize=(10, 10))

# 지역별 총 인구 수 데이터 준비
labels = data['행정구역']
sizes = data['2024년09월_총인구수']

# 파이 차트 그리기
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # 원을 동그랗게 유지
plt.title('지역별 인구 수 비율')
plt.show()
```
