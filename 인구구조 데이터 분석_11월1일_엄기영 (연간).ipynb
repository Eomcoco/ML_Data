# 1.데이터 탐색

## 1.1.데이터 분석을 위한 CSV 파일 로드 & 기본적인 EDA 실행

```python
import pandas as pd
```

```python
data = pd.read_csv('data/202309_202409_주민등록인구및세대현황_월간.csv', encoding='cp949')

data
```

### 1.1.1. 데이터 명칭 간소화

```python
# 행정구역 이름에서 괄호 및 숫자 제거
data['행정구역'] = data['행정구역'].str.replace(r'\s*\(.*?\)', '', regex=True)
```

```python
# '전북특별자치도'와 '전라북도'를 '전북'으로 변경
data['행정구역'] = data['행정구역'].replace({'전북특별자치도': '전북', '전라북도': '전북'})

# 동일한 '전북'을 가진 행을 합쳐서 원본 데이터프레임에 반영
data = data.groupby('행정구역', as_index=False).sum()

```

```python
# 고유한 값 확인
print(data['행정구역'].unique())
```

# 2.인구 구조 시각화

### 2.1. 지역별 총 인구수 

```python
# 2023년 9월 총 인구수 컬럼에서 쉼표 등 특수문자 제거 후 숫자형 변환
data['2023년09월_총인구수'] = data['2023년09월_총인구수'].astype(str).str.replace(r"[^\d]", "", regex=True)
data['2023년09월_총인구수'] = pd.to_numeric(data['2023년09월_총인구수'], errors='coerce')

# 지역별로 그룹화하여 2023년 9월 총 인구수 합산 (전북 통합 포함)
data_grouped = data.groupby('행정구역', as_index=False)['2023년09월_총인구수'].sum()

# 전체 인구수 계산 (2023년 9월 기준)
total_population = data_grouped['2023년09월_총인구수'].sum()

# 각 지역의 인구 비율 계산
data_grouped['인구 비율 (%)'] = (data_grouped['2023년09월_총인구수'] / total_population) * 100

# 결과 확인
data_grouped
```

```python
# 2023년 9월 총 인구수 컬럼에서 쉼표 등 특수문자 제거 후 숫자형 변환
data['2023년09월_총인구수'] = data['2023년09월_총인구수'].astype(str).str.replace(r"[^\d]", "", regex=True)
data['2023년09월_총인구수'] = pd.to_numeric(data['2023년09월_총인구수'], errors='coerce')

# 지역별로 그룹화 없이 2023년 9월 총 인구수와 인구 비율 계산
total_population = data['2023년09월_총인구수'].sum()
data['인구 비율 (%)'] = (data['2023년09월_총인구수'] / total_population) * 100

# 결과 확인
data[['행정구역', '2023년09월_총인구수', '인구 비율 (%)']]
```
